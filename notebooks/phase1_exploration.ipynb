**Clean the Resume Text**

```python
import re

def clean_resume(text):
    
    text = text.lower()
    
    # remove urls
    text = re.sub(r'http\S+', ' ', text)
    
    # remove special characters
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    
    return text.strip()

df['cleaned_text'] = df['Text'].apply(clean_resume)
```
