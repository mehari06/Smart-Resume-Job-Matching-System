```python
import pandas as pd
import re
import string

# Load your dataset
jobs_df = pd.read_csv(r"C:\Users\bezis\Downloads\Smart-Resume-Job-Matching-System\data\job_dataset.csv")  # replace with your path

# Function to clean text
def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()  # lowercase
    text = re.sub(r'\d+', '', text)  # remove numbers
    text = text.translate(str.maketrans('', '', string.punctuation))  # remove punctuation
    text = re.sub(r'\s+', ' ', text)  # remove extra spaces
    text = text.strip()
    return text

# Apply cleaning to relevant columns
jobs_df['Skills_clean'] = jobs_df['Skills'].apply(clean_text)
jobs_df['Keywords_clean'] = jobs_df['Keywords'].apply(clean_text)
jobs_df['Responsibilities_clean'] = jobs_df['Responsibilities'].apply(clean_text)

# Merge Skills + Keywords → requirements
jobs_df['requirements'] = jobs_df['Skills_clean'] + " " + jobs_df['Keywords_clean']

# responsibilities stays as is
jobs_df['responsibilities'] = jobs_df['Responsibilities_clean']

# Combine into one clean_job_text column
jobs_df['clean_job_text'] = "requirements " + jobs_df['requirements'] + " responsibilities " + jobs_df['responsibilities']

# Optional: remove intermediate cleaning columns
jobs_df = jobs_df.drop(columns=['Skills_clean', 'Keywords_clean', 'Responsibilities_clean', 'requirements', 'responsibilities'])

# Check the result
print(jobs_df[['Title', 'clean_job_text']].head())

# Save to CSV for later use
jobs_df.to_csv("cleaned_job_descriptions.csv", index=False)
```
