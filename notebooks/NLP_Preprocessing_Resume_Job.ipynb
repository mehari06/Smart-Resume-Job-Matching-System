**import the essential libraries**

```python
# Data handling
import pandas as pd
import numpy as np

# Text preprocessing
import re
import string

# NLP tools
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK data (only once)
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

**Load Your Cleaned Datasets**

```python
# Load your cleaned resume dataset
resume_df = pd.read_csv(r"C:\Users\bezis\Downloads\Smart-Resume-Job-Matching-System\data\tech_resumes_filtered.csv")  # replace with your path
print("Resume Dataset Shape:", resume_df.shape)
resume_df.head()

# Load your cleaned job description dataset
job_df = pd.read_csv(r"C:\Users\bezis\Downloads\Smart-Resume-Job-Matching-System\data\cleaned_job_descriptions.csv")  # replace with your path
print("Job Description Dataset Shape:", job_df.shape)
job_df.head()
```

**Define Preprocessing Function**

```python
# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # Lowercase
    text = text.lower()
    
    # Remove punctuation & numbers
    text = re.sub(r'[^a-z\s]', ' ', text)
    
    # Tokenize
    tokens = text.split()
    
    # Remove stopwords and lemmatize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    
    # Rejoin tokens
    return ' '.join(tokens)
```

```python
# Apply to resume dataset
resume_df['preprocessed_text'] = resume_df['cleaned_text'].apply(preprocess_text)

# Apply to job description dataset
job_df['preprocessed_text'] = job_df['clean_job_text'].apply(preprocess_text)

# Inspect first few rows
resume_df[['cleaned_text', 'preprocessed_text']].head()
job_df[['clean_job_text', 'preprocessed_text']].head()
```

```python
# Save preprocessed resumes
resume_df.to_csv("preprocessed_resumes.csv", index=False)

# Save preprocessed job descriptions
job_df.to_csv("preprocessed_jobs.csv", index=False)
```
